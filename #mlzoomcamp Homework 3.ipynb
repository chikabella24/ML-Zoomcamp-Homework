Question 6

For this question, we'll see how to use a linear regression model from Scikit-Learn

We'll need to use the original column 'median_house_value'. Apply the logarithmic transformation to this column.

Fit the Ridge regression model (model = Ridge(alpha=a, solver="sag", random_state=42)) on the training data.

This model has a parameter alpha. Let's try the following values: [0, 0.01, 0.1, 1, 10]
Which of these alphas leads to the best RMSE on the validation set? Round your RMSE scores to 3 decimal digits.

If there are multiple options, select the smallest alpha.

```python
#Applying the logarithnmic transformation to the median_house_value
df['median_house_value']=np.log1p(df['median_house_value'])
```

```python
#splitting the data into full_train and test with a random state 42
# and also splitting the full_train into train and validitation
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
# getting the y values
y_train = df_train.median_house_value.values
y_val = df_val.median_house_value.values
y_test = df_test.median_house_value.values
```

```python
#deleting the target from the train,test and val
del df_train["median_house_value"]
del df_val["median_house_value"]
del df_test["median_house_value"]
```

```python
#preparing data for training and testing
train_dicts =df_train.to_dict(orient = 'records')
dv = DictVectorizer(sparse = False)
dv.fit(train_dicts)
```

```python
X_train = dv.fit_transform(train_dicts)
```

```python
val_dict = df_val.to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
mse = mean_squared_error

alpha_list = [0, 0.01, 0.1, 1, 10]

for a in alpha_list:
    model = Ridge(alpha=a, solver="sag", random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    
    score = np.sqrt(mse(y_pred, y_val))
        
    print(f"The score for alpha parameter {a} -- {round(score,3)}")
```
